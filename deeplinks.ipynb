```python
import astropy.units as u
from astropy.time import Time
from astropy.coordinates import SkyCoord, AltAz, EarthLocation
import numpy as np
```

```python
# Site and time
lat, lon, height_m = 51.445797, 7.262352, 141 # Bochum, Germany
location = EarthLocation(lat=lat*u.deg, lon=lon*u.deg, height=height_m*u.m)
t_obs = Time("2026-03-04T20:30:00", scale="utc")

# Scan definition
altitudes_deg = [15, 30, 45, 60, 75]
az_start, az_end = 0.0, 360.0 
az_list = np.arange(az_start, az_end, 45.0)

pointings = []
for alt_deg in altitudes_deg:
    for az_deg in az_list:
        altaz = SkyCoord(
            alt=alt_deg*u.deg,
            az=az_deg*u.deg,
            frame=AltAz(obstime=t_obs, location=location)
        )
        icrs = altaz.icrs  # gives RA/Dec
        pointings.append({
            "alt_deg": alt_deg,
            "az_deg": az_deg,
            "ra_deg": icrs.ra.deg,
            "dec_deg": icrs.dec.deg
        })

for p in pointings:
    print(f"Alt={p['alt_deg']:>5.1f} Az={p['az_deg']:>6.2f} -> "
          f"RA={p['ra_deg']:>8.3f} Dec={p['dec_deg']:>8.3f}")
```

```python
base_tail = "c=3970&et=3970&g=20&d=12&t=1772663990"

# Group pointings by altitude (in creation order)
from collections import defaultdict
by_alt = defaultdict(list)
for p in pointings:
    by_alt[p["alt_deg"]].append(p)

all_links = []
counter = 1

for alt in sorted(by_alt.keys()):
    print(f"* {int(alt)} degrees")

    ring = by_alt[alt]
    ring8 = ring[:7] + [ring[0]]

    for p in ring8:
        group_id = f"x20260304skyglow{counter:05d}"
        link = (
            f"unistellar://science/transient?"
            f"ra={p['ra_deg']:.6f}&dec={p['dec_deg']:.6f}&{base_tail}&scitag={group_id}"
        )

        test_name = f"Test {counter}_alt_{int(alt)}"

        # HTML block for every link
        print('[[html]]')
        print(f'<a href="{link}"> {test_name} </a>')
        print('[[/html]]')

        all_links.append(link)
        counter += 1
```
